```python
features = dict()
with open("SDA_full.tsv", "r") as inp:
    for line in inp.readlines()[1:]:
        line = line.strip().split("\t")
        name = line[1]
        #print(line[5:])
        features[name] = set(line[5:])
```

```python
with open("kegg_sim.tsv", "w") as oup:
    print >> oup, "A\tB\tsim"
    for x in features:
        for y in features:
            if x == y: continue
            sim = len(features[x]&features[y])*1./len(features[x]|features[y])
            if sim > 0.1:
                print >> oup, "\t".join([x, y, str(sim)])
```

```python
features = dict()
with open("SDA.tsv", "r") as inp:
    for line in inp.readlines()[1:]:
        line = line.strip().split("\t")
        name = line[1]
        #print(line[5:])
        features[name] = set(line[5:])
```

```python
from collections import Counter
genes = Counter()
for x in features:
    genes += Counter(features[x])
```

```python
gene_name = set([g.split("(")[0] for g in genes])
print(len(gene_name))
```

```python
top_genes = genes.most_common()
```

```python
top_genes
```

```python
RNAseq = set([x[0].split("(")[0] for x in top_genes if "RNAseq" in x[0]])
Meth = set([x[0].split("(")[0] for x in top_genes if "Methyl" in x[0]])
```

```python
len(Meth)
```

```python
import glob
with open("SDA_GGI.tsv", "w") as oup:
    print >> oup, "A\tB\tRel"
    for x in glob.glob("kegg_db/*_edges.txt"):
        with open(x, "r") as inp:
            for line in inp.read().strip().split("\n"):
                if len(line) == 0: continue
                a, b, tp = line.strip().split()
                if a in RNAseq and b in RNAseq: 
                    print >> oup, "\t".join([a, b, tp])
                    #print "\t".join([a, b, tp])
```

```python
with open("SDA_Meth.tsv", "w") as oup:
    print >> oup, "gene\tMeth"
    for meth in Meth:
        print >> oup, meth + "\t" + "Meth"
```

```python
from collections import defaultdict
kegg_set = defaultdict(set)
for x in RNAseq:
    for z in features:
        if x + "(RNAseq)" in features[z]:
            kegg_set[x].add(z)
```

```python
with open("SDA_kegg.tsv", "w") as oup:
    print >> oup, "gene\tkegg"
    for x in kegg_set:
        y = kegg_set[x]
        y = ",".join(y)
        print >> oup, (x + "\t" + y)
```
